```python
import pandas as pd
import numpy as np
```

In this notebook, we create a simple dataset containing city names and their populations. The `City` column currently stores full city names. 

Our objective is to transform these categorical values into standardized city codes 
(e.g., Istanbul â†’ IST) using a mapping approach.

```python
df = pd.DataFrame({'City': ['Istanbul', 'Ankara', 'Izmir', 'Bursa', 'Antalya'],
                   'Population': [15519267, 5639076, 4470587, 3101833, 2619832]})

df
```

Let's create a mapping dictionary for each city name and apply it to the dataset.

```python
city_mapping = {'Istanbul': 'IST',
                'Ankara': 'ANK',
                'Izmir': 'IZM',
                'Bursa': 'BUR',
                'Antalya': 'ANT'}

df['City'] = df['City'].map(city_mapping)

df
```
